# Analysing the forecasts - Version 2

```python
# ============================================================================
# EPIDEMIC FORECASTING ANALYSIS - FINAL WEEK ONLY
# Compare 1-week, 2-week, 4-week, 8-week forecasts (last week only)
# ============================================================================

# CELL 1: Imports and Setup
# ============================================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (15, 8)
plt.rcParams['font.size'] = 10

# ============================================================================
# CELL 2: Load Results and Extract FINAL WEEK Only
# ============================================================================

horizons = ['1week', '2week', '4week', '8week']
results_dir = Path('./results')

# Load and extract FINAL week predictions only
final_week_results = {}

for horizon in horizons:
    result_path = results_dir / f'sentinelle_ILI_{horizon}'
    
    # Load predictions
    preds_full = np.load(result_path / 'preds_original.npy')
    trues_full = np.load(result_path / 'trues_original.npy')
    
    # Handle 1D arrays
    if preds_full.ndim == 1:
        preds_full = preds_full.reshape(-1, 1)
        trues_full = trues_full.reshape(-1, 1)
    
    # Extract ONLY the final week (last column)
    final_week_idx = preds_full.shape[1] - 1  # Last column index
    
    final_week_results[horizon] = {
        'preds': preds_full[:, final_week_idx],  # 1D array of final week predictions
        'trues': trues_full[:, final_week_idx],  # 1D array of final week true values
        'horizon_weeks': preds_full.shape[1]      # How many weeks ahead this model predicts
    }
    
    print(f"{horizon}: Extracted week {final_week_idx + 1} predictions (shape: {final_week_results[horizon]['preds'].shape})")

# ============================================================================
# CELL 3: Compute Metrics for Final Week Only
# ============================================================================

def compute_metrics(preds, trues):
    """Compute MAE, MSE, RMSE, MAPE"""
    mae = np.mean(np.abs(preds - trues))
    mse = np.mean((preds - trues)**2)
    rmse = np.sqrt(mse)
    # Avoid division by zero
    mape = np.mean(np.abs((preds - trues) / np.maximum(trues, 1))) * 100
    return {'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'MAPE': mape}

# Compute metrics
metrics_data = []

for horizon in horizons:
    preds = final_week_results[horizon]['preds']
    trues = final_week_results[horizon]['trues']
    weeks_ahead = final_week_results[horizon]['horizon_weeks']
    
    metrics = compute_metrics(preds, trues)
    
    metrics_data.append({
        'Forecast Horizon': f'{weeks_ahead}-week ahead',
        'MAE (cases)': f"{metrics['MAE']:.0f}",
        'RMSE (cases)': f"{metrics['RMSE']:.0f}",
        'MAPE (%)': f"{metrics['MAPE']:.1f}",
        'Mean True Value': f"{trues.mean():.0f}",
        'MAE / Mean': f"{metrics['MAE'] / trues.mean():.2f}x"
    })

df_metrics = pd.DataFrame(metrics_data)

print("\n" + "="*100)
print("FINAL WEEK FORECAST PERFORMANCE")
print("(Only evaluating the furthest week ahead for each model)")
print("="*100)
print(df_metrics.to_string(index=False))
print("="*100)

# ============================================================================
# CELL 4: Comparison Bar Chart
# ============================================================================

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

horizon_labels = ['1-week', '2-week', '4-week', '8-week']
colors = ['#3498db', '#2ecc71', '#e74c3c', '#f39c12']

# Extract numeric values
maes = [float(m['MAE (cases)'].replace(',', '')) for m in metrics_data]
rmses = [float(m['RMSE (cases)'].replace(',', '')) for m in metrics_data]
mapes = [float(m['MAPE (%)']) for m in metrics_data]

# MAE Plot
axes[0].bar(horizon_labels, maes, color=colors, alpha=0.8, edgecolor='black', linewidth=2)
axes[0].set_ylabel('MAE (Cases)', fontsize=12, fontweight='bold')
axes[0].set_title('Mean Absolute Error', fontsize=14, fontweight='bold')
axes[0].grid(axis='y', alpha=0.3)
for i, mae in enumerate(maes):
    axes[0].text(i, mae, f'{mae:.0f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# RMSE Plot
axes[1].bar(horizon_labels, rmses, color=colors, alpha=0.8, edgecolor='black', linewidth=2)
axes[1].set_ylabel('RMSE (Cases)', fontsize=12, fontweight='bold')
axes[1].set_title('Root Mean Squared Error', fontsize=14, fontweight='bold')
axes[1].grid(axis='y', alpha=0.3)
for i, rmse in enumerate(rmses):
    axes[1].text(i, rmse, f'{rmse:.0f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# MAPE Plot
axes[2].bar(horizon_labels, mapes, color=colors, alpha=0.8, edgecolor='black', linewidth=2)
axes[2].set_ylabel('MAPE (%)', fontsize=12, fontweight='bold')
axes[2].set_title('Mean Absolute Percentage Error', fontsize=14, fontweight='bold')
axes[2].grid(axis='y', alpha=0.3)
for i, mape in enumerate(mapes):
    axes[2].text(i, mape, f'{mape:.1f}%', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.suptitle('Final Week Forecast Performance Comparison', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('final_week_metrics_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# CELL 5: Time Series Plot with Dates - Final Week Only
# ============================================================================

def plot_final_week_timeseries(horizon):
    """Plot final week predictions vs true values with dates"""
    
    preds = final_week_results[horizon]['preds']
    trues = final_week_results[horizon]['trues']
    weeks_ahead = final_week_results[horizon]['horizon_weeks']
    
    # Load dates
    df_raw = pd.read_csv('./epidemics_30years_full.csv')
    df_raw['date'] = pd.to_datetime(df_raw['date'])
    
    # Calculate test dates
    total_len = len(df_raw)
    num_test = int(total_len * 0.2)
    test_start_idx = total_len - num_test
    
    seq_len = 26
    # For final week, add the full prediction length
    forecast_start_idx = test_start_idx + seq_len + (weeks_ahead - 1)
    
    num_samples = len(preds)
    test_dates = df_raw['date'].iloc[forecast_start_idx:forecast_start_idx + num_samples].values
    
    # Truncate to match
    min_len = min(len(test_dates), len(preds))
    test_dates = test_dates[:min_len]
    preds = preds[:min_len]
    trues = trues[:min_len]
    
    # Compute metrics
    mae = np.mean(np.abs(preds - trues))
    rmse = np.sqrt(np.mean((preds - trues)**2))
    
    # Plot
    fig, ax = plt.subplots(figsize=(20, 7))
    
    ax.plot(test_dates, trues, 'o-', color='green', linewidth=2, 
            markersize=4, label='True Values', alpha=0.7)
    ax.plot(test_dates, preds, 's-', color='red', linewidth=2, 
            markersize=4, label='Predictions', alpha=0.7)
    ax.fill_between(test_dates, preds, trues, alpha=0.2, color='gray')
    
    ax.set_xlabel('Date', fontsize=13, fontweight='bold')
    ax.set_ylabel('Epidemic Cases (I_child)', fontsize=13, fontweight='bold')
    ax.set_title(f'{weeks_ahead}-Week Ahead Forecast | MAE: {mae:.0f} | RMSE: {rmse:.0f}', 
                 fontsize=15, fontweight='bold')
    ax.legend(loc='best', fontsize=12)
    ax.grid(alpha=0.3)
    
    # Format dates
    import matplotlib.dates as mdates
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    ax.xaxis.set_major_locator(mdates.MonthLocator(interval=6))
    plt.xticks(rotation=45, ha='right')
    
    # Date range info
    date_range = f"{test_dates[0]} to {test_dates[-1]}"
    ax.text(0.02, 0.98, f'Test Period: {date_range}\nForecasts: {len(preds)}', 
            transform=ax.transAxes, fontsize=10, 
            verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    plt.tight_layout()
    plt.savefig(f'timeseries_final_{horizon}.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print(f"{horizon}: {weeks_ahead}-week ahead forecast")
    print(f"  Period: {test_dates[0]} to {test_dates[-1]}")
    print(f"  MAE: {mae:.0f} cases | RMSE: {rmse:.0f} cases")
    print(f"  Mean true value: {trues.mean():.0f} cases\n")

# Plot all horizons
for horizon in horizons:
    plot_final_week_timeseries(horizon)

# ============================================================================
# CELL 6: Scatter Plot - Predictions vs True
# ============================================================================

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for idx, horizon in enumerate(horizons):
    preds = final_week_results[horizon]['preds']
    trues = final_week_results[horizon]['trues']
    weeks_ahead = final_week_results[horizon]['horizon_weeks']
    
    ax = axes[idx]
    
    # Scatter
    ax.scatter(trues, preds, alpha=0.4, s=30, color='steelblue', edgecolor='black', linewidth=0.5)
    
    # Perfect prediction line
    min_val = min(trues.min(), preds.min())
    max_val = max(trues.max(), preds.max())
    ax.plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2, label='Perfect Prediction')
    
    # R
    correlation = np.corrcoef(trues, preds)[0, 1]
    r2 = correlation ** 2
    
    # MAE
    mae = np.mean(np.abs(preds - trues))
    
    ax.set_xlabel('True Values (Cases)', fontsize=12, fontweight='bold')
    ax.set_ylabel('Predicted Values (Cases)', fontsize=12, fontweight='bold')
    ax.set_title(f'{weeks_ahead}-Week Ahead | R={r2:.3f} | MAE={mae:.0f}', 
                 fontsize=13, fontweight='bold')
    ax.legend(fontsize=10)
    ax.grid(alpha=0.3)

plt.suptitle('Predictions vs True Values - Final Week Only', fontsize=16, fontweight='bold', y=1.00)
plt.tight_layout()
plt.savefig('scatter_final_week.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# CELL 7: Error Distribution
# ============================================================================

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for idx, horizon in enumerate(horizons):
    preds = final_week_results[horizon]['preds']
    trues = final_week_results[horizon]['trues']
    weeks_ahead = final_week_results[horizon]['horizon_weeks']
    
    errors = preds - trues
    
    ax = axes[idx]
    ax.hist(errors, bins=50, color='steelblue', alpha=0.7, edgecolor='black')
    ax.axvline(0, color='red', linestyle='--', linewidth=2, label='Zero Error')
    ax.axvline(errors.mean(), color='orange', linestyle='--', linewidth=2, 
               label=f'Mean Error: {errors.mean():.0f}')
    
    ax.set_xlabel('Prediction Error (Cases)', fontsize=12, fontweight='bold')
    ax.set_ylabel('Frequency', fontsize=12, fontweight='bold')
    ax.set_title(f'{weeks_ahead}-Week Ahead - Error Distribution', fontsize=13, fontweight='bold')
    ax.legend(fontsize=10)
    ax.grid(axis='y', alpha=0.3)

plt.suptitle('Error Distribution - Final Week Only', fontsize=16, fontweight='bold', y=1.00)
plt.tight_layout()
plt.savefig('error_distribution_final.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# CELL 8: Summary Statistics Table
# ============================================================================

summary_stats = []

for horizon in horizons:
    preds = final_week_results[horizon]['preds']
    trues = final_week_results[horizon]['trues']
    weeks_ahead = final_week_results[horizon]['horizon_weeks']
    
    errors = preds - trues
    
    summary_stats.append({
        'Horizon': f'{weeks_ahead}-week',
        'MAE': f"{np.mean(np.abs(errors)):.0f}",
        'RMSE': f"{np.sqrt(np.mean(errors**2)):.0f}",
        'Mean Error': f"{errors.mean():.0f}",
        'Std Error': f"{errors.std():.0f}",
        'Min Error': f"{errors.min():.0f}",
        'Max Error': f"{errors.max():.0f}",
        'Mean True': f"{trues.mean():.0f}"
    })

df_summary = pd.DataFrame(summary_stats)

print("\n" + "="*120)
print("DETAILED SUMMARY STATISTICS - FINAL WEEK FORECASTS ONLY")
print("="*120)
print(df_summary.to_string(index=False))
print("="*120)

print("\n Analysis complete! All plots saved.")
print("\nGenerated files:")
print("  - final_week_metrics_comparison.png")
print("  - timeseries_final_*.png (4 files)")
print("  - scatter_final_week.png")
print("  - error_distribution_final.png")
```
